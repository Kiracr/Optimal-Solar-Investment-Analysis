```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
benin = pd.read_csv("/Users/Suraphel/Desktop/Week-0/Optimal-Solar-Investment-Analysis/notebooks/benin.csv")
sierraleone = pd.read_csv("/Users/Suraphel/Desktop/Week-0/Optimal-Solar-Investment-Analysis/notebooks/sierraleone.csv")
togo = pd.read_csv("/Users/Suraphel/Desktop/Week-0/Optimal-Solar-Investment-Analysis/notebooks/togo.csv")
```

```python

def plot_wind_rose(data, title):
    # Define wind directions (0-360 degrees) and corresponding labels
    wind_directions = np.array([0, 45, 90, 135, 180, 225, 270, 315])
    wind_speed_means = []
    
    # Calculate mean wind speed for each direction
    for direction in wind_directions:
        # Select wind speed for each direction within a small range around it
        mask = (data['WD'] >= direction - 22.5) & (data['WD'] < direction + 22.5)
        mean_wind_speed = data[mask]['WS'].mean()
        wind_speed_means.append(mean_wind_speed)
    
    # Convert to radians for polar plotting
    angles = np.deg2rad(wind_directions)

    # Plotting wind rose using polar coordinates
    fig, ax = plt.subplots(subplot_kw={'projection': 'polar'}, figsize=(8, 8))
    
    # Create the bars for the wind rose
    ax.bar(angles, wind_speed_means, width=np.pi/4, color='skyblue', edgecolor='black', alpha=0.7)

    # Adjust the wind rose plot appearance
    ax.set_theta_zero_location('N')
    ax.set_theta_direction(-1)
    ax.set_rticks([0, max(wind_speed_means)/2, max(wind_speed_means)])  # Set radial ticks
    ax.set_rlabel_position(90)  # Position the radial labels

    # Title and display
    plt.title(title)
    plt.show()


```

```python
# plot wind rose for benin 
plot_wind_rose(benin, 'Wind Rose for Benin - Malanville')
```

```python
# plot wind rose for Sierraleone 
plot_wind_rose(sierraleone, 'Wind Rose for Sierraleone - bumbuna')
```

```python
# plot wind rose for togo 
plot_wind_rose(togo, 'Wind Rose for Togo - dapaong')
```

## Conclusion

The wind analysis using the wind rose method successfully visualized the distribution of wind speed across different directions for the datasets from Benin, Sierra Leone, and Togo. The analysis provided insights into wind patterns and variability in each location, helping identify significant wind events and trends in wind direction.
