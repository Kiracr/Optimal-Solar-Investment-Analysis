## Histograms

```python
# import library and data
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np
 
```

```python
benin = pd.read_csv('C:/Users/Suraphel/Desktop/Week-0/Optimal-Solar-Investment-Analysis/notebooks/benin.csv')
sierraleone = pd.read_csv('C:/Users/Suraphel/Desktop/Week-0/Optimal-Solar-Investment-Analysis/notebooks/sierraleone.csv')
togo = pd.read_csv('C:/Users/Suraphel/Desktop/Week-0/Optimal-Solar-Investment-Analysis/notebooks/togo.csv')
```

```python
variables = ['GHI', 'DNI', 'DHI', 'WS', 'Tamb']


for variable in variables:
    plt.figure(figsize=(10, 6))
    
    # Plot for Benin
    plt.hist(benin[variable].dropna(), bins=30, alpha=0.5, label='Benin')
    
    # Plot for Sierra Leone
    plt.hist(sierraleone[variable].dropna(), bins=30, alpha=0.5, label='Sierra Leone')
    
    # Plot for Togo
    plt.hist(togo[variable].dropna(), bins=30, alpha=0.5, label='Togo')
    
    plt.title(f'Histogram of {variable}')
    plt.xlabel(variable)
    plt.ylabel('Frequency')
    plt.legend(loc='upper right')
    plt.grid(True)
    plt.show()
```

## Bubble Chart

```python
# Bubble chart for GHI vs. Tamb vs. WS with bubble size representing RH

# Define the variables for the bubble chart
x_variable = 'GHI'
y_variable = 'Tamb'
bubble_size_variable = 'WS'
bubble_color_variable = 'RH'

# Plot for Benin
plt.figure(figsize=(10, 6))
plt.scatter(benin[x_variable], benin[y_variable], s=benin[bubble_size_variable], c=benin[bubble_color_variable], alpha=0.5, cmap='viridis')
plt.title(f'Bubble Chart of {x_variable} vs. {y_variable} vs. {bubble_size_variable} (Benin)')
plt.xlabel(x_variable)
plt.ylabel(y_variable)
plt.colorbar(label=bubble_color_variable)
plt.grid(True)
plt.show()

# Plot for Sierra Leone
plt.figure(figsize=(10, 6))
plt.scatter(sierraleone[x_variable], sierraleone[y_variable], s=sierraleone[bubble_size_variable], c=sierraleone[bubble_color_variable], alpha=0.5, cmap='viridis')
plt.title(f'Bubble Chart of {x_variable} vs. {y_variable} vs. {bubble_size_variable} (Sierra Leone)')
plt.xlabel(x_variable)
plt.ylabel(y_variable)
plt.colorbar(label=bubble_color_variable)
plt.grid(True)
plt.show()

# Plot for Togo
plt.figure(figsize=(10, 6))
plt.scatter(togo[x_variable], togo[y_variable], s=togo[bubble_size_variable], c=togo[bubble_color_variable], alpha=0.5, cmap='viridis')
plt.title(f'Bubble Chart of {x_variable} vs. {y_variable} vs. {bubble_size_variable} (Togo)')
plt.xlabel(x_variable)
plt.ylabel(y_variable)
plt.colorbar(label=bubble_color_variable)
plt.grid(True)
plt.show()
```
